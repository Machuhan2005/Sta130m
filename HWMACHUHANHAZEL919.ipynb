```python
#Question1:

#Example in real life scenario
#Setup: There are three doors:

#Door 1: Goat
#Door 2: Car
#Door 3: Goat

#Initial Choice: The contestant randomly chooses Door 1.

#Host's Action: The host knows where the car is and opens one of the remaining doors. Since Alice chose Door 1, the host opens Door 3, which has a goat behind it.

#Final Decision: Now, Alice has the option to either stick with her initial choice (Door 1) or switch to Door 2, which is the only unopened door.

#If Alice sticks with Door 1, she wins a goat.
#If Alice switches to Door 2, she wins the car.

import numpy as np  # Importing the numpy library for random choices

# Initialization of the game setup
all_door_options = (1, 2, 3)  # Tuple representing the three doors
my_door_choice = 1  # Contestant's initial choice (arbitrarily set to door 1)
i_won = 0  # Counter to keep track of wins
reps = 100000  # Number of simulations to run for statistical accuracy

# Run the simulation for the specified number of repetitions
for _ in range(reps):  # Use '_' since the loop variable isn't needed
    # Randomly select which door has the car behind it
    secret_winning_door = np.random.choice(all_door_options)

    # Create a list of doors excluding the contestant's initial choice
    remaining_doors = list(all_door_options)  # Convert tuple to a list for manipulation
    remaining_doors.remove(my_door_choice)  # Remove the contestant's initial choice from the list

    # The host reveals one of the remaining doors that has a goat
    goat_door_reveal = np.random.choice([door for door in remaining_doors if door != secret_winning_door])
    remaining_doors.remove(goat_door_reveal)  # Remove the revealed goat door from the list

    # The contestant switches to the only remaining door
    my_door_choice = remaining_doors[0]  # Only one door left to choose

    # Check if the contestant's final choice is the winning door
    if my_door_choice == secret_winning_door:
        i_won += 1  # Increment the win counter if the contestant wins

# Calculate the win rate based on the number of wins
win_rate = i_won / reps
print(f"Win rate by switching: {win_rate:.4f}")  # Print the win rate formatted to four decimal places
```

```python
#Question2 answer: simpler, more streamlined code for this question

import numpy as np  # Importing numpy for random choices

# Initialization
num_simulations = 100000  # Number of simulations to run
wins = 0  # Counter for wins

# Run the simulation for the specified number of repetitions
for _ in range(num_simulations):
    # Randomly assign the winning door
    winning_door = np.random.choice([1, 2, 3])
    # Contestant makes an initial choice
    contestant_choice = np.random.choice([1, 2, 3])

    # Host reveals a goat door
    doors = [1, 2, 3]
    doors.remove(contestant_choice)  # Remove contestant's choice
    doors.remove(winning_door if winning_door in doors else doors[0])  # Remove one goat door
    # Switch to the remaining door
    contestant_choice = [door for door in [1, 2, 3] if door not in [contestant_choice] + doors][0]

    # Check if the contestant wins
    if contestant_choice == winning_door:
        wins += 1  # Increment win counter

# Calculate and print the win rate
win_rate = wins / num_simulations
print(f"Win rate by switching: {win_rate:.4f}")  # Print the win rate formatted to four decimal places

```
