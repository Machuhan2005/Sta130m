Question5:According to GünTürkün (2003), the goal is to test whether couples tilt their heads randomly when kissing, with a 50/50 chance of tilting left or right (the null hypothesis \(H_0\)). Out of 124 couples, 80 (about 64.5%) tilted their heads to the right. I assume the null hypothesis is that couples tilt their heads to the left or right with equal probability, meaning a 50/50 distribution. In other words, if H0 holds, each couple has a 50% chance of tilting their heads to the right. However, in this study, I observed that 80 out of 124 couples (around 64.5%) tilted their heads to the right. To determine if this result could be due to random chance, I conducted a hypothesis test using simulation. I repeated the experiment 10,000 times, with each simulation assuming a 50% chance of tilting right, and I counted how often the number of couples tilting right was greater than or equal to 80. The resulting p-value was 0.0013. Based on the given criteria, since 0.01 ≥ p > 0.001, I conclude that there is strong evidence against the null hypothesis. Therefore, I conclude that couples indeed tend to tilt their heads to the right while kissing, and this tendency is not due to randomness.

```python
import numpy as np

n_simulations = 10000
n_couples = 124
observed_right_tilt = 80

simulations = np.random.binomial(n=n_couples, p=0.5, size=n_simulations)
p_value = np.mean(simulations >= observed_right_tilt)

print(f"P-value: {p_value}")
```

Q: Does a small p-value actually prove the null hypothesis is false?
A: Not really. A small p-value just tells us that the data we got is pretty unlikely if the null hypothesis were true, so it makes sense to reject the null. But it doesn’t prove it’s wrong—it could just mean we saw a rare event by chance. We use decision rules like p < 0.05 to determine whether the evidence against the null hypothesis is strong enough, but even that doesn’t guarantee we’ve made the right conclusion.

Q: Can we ever say with 100% certainty that Fido is innocent just by looking at a p-value?
A: We are not 100% certain even if we get a very small p-value. This is because statistical tests are prone to two types of errors. Type I error (false positive) means we mistakenly reject Fido's innocence when he’s actually innocent. Type II error (false negative) means we fail to reject his innocence when he’s actually guilty. A large p-value just means we don't have enough evidence to say he’s guilty, but it doesn’t confirm his innocence—it’s more like saying, “We don’t have a strong case against him.”

Q: Can a p-value ever fully prove that Fido is guilty?
A: Nope. Even if the p-value is super small, it just tells us the data doesn’t fit with the idea that Fido is innocent. But we can’t say for sure he’s guilty—there’s always a chance we’re wrong, called a Type I error (false positive).

Q: How small or big does a p-value have to be for us to say something is definitely true or false?
A: We can’t ever be 100% sure with just a p-value. People usually use 0.05 as a cutoff—if p < 0.05, they reject the null hypothesis. But in some fields, like medicine, they might want stricter cutoffs like 0.01 or 0.001 to avoid mistakes. Even then, there’s always some risk of being wrong, so p-values only guide us—they don’t give us absolute truth.

```python
import numpy as np
import pandas as pd
import plotly.figure_factory as ff
import plotly.graph_objects as go


np.random.seed(42)

n_size = 10 
number_of_simulations = 10000  
population_parameter_value_under_H0 = 0.5  

patient_data = pd.DataFrame({
    'HealthScoreChange': np.random.randint(-10, 10, size=n_size)
})


observed_statistic = (patient_data.HealthScoreChange > 0).mean()


IncreaseProportionSimulations_underH0random = np.zeros(number_of_simulations)
for i in range(number_of_simulations):
    random_improvement = np.random.choice([0, 1], size=n_size, replace=True)
    IncreaseProportionSimulations_underH0random[i] = random_improvement.mean()


if observed_statistic >= population_parameter_value_under_H0:

    SimStats_one_tailed = (
        IncreaseProportionSimulations_underH0random >= observed_statistic
    )
else:

    SimStats_one_tailed = (
        IncreaseProportionSimulations_underH0random <= observed_statistic
    )


p_value_one_tailed = SimStats_one_tailed.sum() / number_of_simulations


print(f"One-Tailed Test p-value: {p_value_one_tailed}")


hist_data = [IncreaseProportionSimulations_underH0random + 
             np.random.uniform(-0.05, 0.05, size=len(IncreaseProportionSimulations_underH0random))]
group_labels = ['SYNTHETICALLY SIMULATED Sampling Distribution<br>under H0']

fig = ff.create_distplot(hist_data, group_labels, curve_type='normal',
                         show_hist=True, show_rug=False, bin_size=0.1)


pv_y = 2.5
pv_y_ = 0.25
fig.add_shape(type="line", x0=observed_statistic, y0=0,
              x1=observed_statistic, y1=pv_y,
              line=dict(color="Green", width=4), name="Observed Statistic")

fig.add_trace(go.Scatter(x=[observed_statistic], y=[pv_y + pv_y_],
                         text=["Observed<br>Statistic<br>^"], mode="text", showlegend=False))


if observed_statistic >= population_parameter_value_under_H0:

    fig.add_shape(type="rect", x0=observed_statistic, y0=0, x1=1.25, y1=pv_y,
                  fillcolor="LightCoral", opacity=0.5, line_width=0)
else:
 
    fig.add_shape(type="rect", x0=-0.25, y0=0, x1=observed_statistic, y1=pv_y,
                  fillcolor="LightCoral", opacity=0.5, line_width=0)


fig.update_layout(
    title="One-Tailed Test: SYNTHETICALLY SIMULATED Sampling Distribution<br>under H0 with p-value region",
    xaxis_title="Mean Health Score Change", yaxis_title="Density", yaxis=dict(range=[0, pv_y + 2 * pv_y_])
)


fig.show()
```

In this hypothesis test, I want to determine whether students can tell if tea or milk was poured first, or if their answers are just random guesses. The null hypothesis (H₀) assumes the chance of correctly identifying the order is 50%, meaning the results are purely by chance. The alternative hypothesis (H₁) suggests that the students’ answers are not random, showing some ability to tell the difference. From my sample, the observed statistic is 49/80 = 0.6125.

To test this, I used simulation to generate a sampling distribution, assuming under H₀ that the probability of a correct answer is 50%. I ran 10,000 simulations, where 80 students guessed randomly each time, and I calculated the proportion of correct answers. I then counted how often the simulated proportion was at least as large as the observed 0.6125. The p-value represents the probability of getting this result or something more extreme if the null hypothesis is true.

If the p-value is below 0.05, I reject the null hypothesis, suggesting the students may have some real ability to tell the difference. If it’s 0.05 or higher, I fail to reject the null, meaning the results could just be due to chance. This analysis gives me a clearer understanding of the data and hypotheses. Here’s the code I used:

```python
import numpy as np

np.random.seed(42)

n_students = 80  
observed_correct = 49  
p_null = 0.5  
n_simulations = 10000  

phat_observed = observed_correct / n_students

simulated_statistics = np.random.binomial(n_students, p_null, n_simulations) / n_students

p_value = np.mean(simulated_statistics >= phat_observed)

print(f"Observed Statistic: {phat_observed}")
print(f"p-value: {p_value}")

if p_value < 0.05:
    print("p-value < 0.05, reject the null hypothesis. Students may have some ability to judge correctly.")
else:
    print("p-value >= 0.05, fail to reject the null hypothesis. The results may be due to random guessing.")
```

This code simulates 10,000 trials, calculating the proportion of correct answers in each trial assuming random guessing. The p-value reflects how often the simulated results were at least as extreme as what I observed. If the p-value is small, it suggests the students’ answers are unlikely to be random, indicating some ability. If it’s large, I can’t reject the idea that their answers were just guesses.
